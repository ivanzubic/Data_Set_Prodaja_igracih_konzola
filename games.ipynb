```python
missing = games.isna().sum()
print(missing)
missing.plot(kind='bar', figsize=(8,4), title='Broj nedostajućih vrijednosti po stupcu')
plt.ylabel('Broj NaN vrijednosti')
plt.xticks(rotation=45)
plt.show()
```

```python
dups = games[games.duplicated()]
print('Duplikati:')
display(dups)

games = games.drop_duplicates()
print(f'Broj redaka nakon uklanjanja duplikata: {len(games)}')
```

```python
games['Year'] = games['Year'].fillna(games['Year'].median())
#df['solar_kw'] = df['solar_kw'].fillna(df['solar_kw'].median())
#df['temperature_avg'] = df['temperature_avg'].fillna(df['temperature_avg'].mean())

print(games.isna().sum())
```

```python
sales_per_year = games.groupby('Year')['EU_Sales'].sum()

print(sales_per_year.head())
plt.figure()
plt.plot(sales_per_year.index, sales_per_year.values)

plt.xlabel("Godina")
plt.ylabel("Prodaja u Europi (milijuni)")
plt.title("Prodaja video igara po godinama")

plt.show()
```

```python
platform_sales =games.groupby('Platform')['EU_Sales'].sum()
platform_sales.sort_values().tail(10).plot(kind='bar')
plt.show()
```

```python
platform_sales =games.groupby('Name')['EU_Sales'].sum()
platform_sales.sort_values().tail(10).plot(kind='bar')
plt.show()
```

```python
platform_sales =games.groupby('Genre')['EU_Sales'].sum()
platform_sales.sort_values().tail(10).plot(kind='bar')
plt.show()
```
